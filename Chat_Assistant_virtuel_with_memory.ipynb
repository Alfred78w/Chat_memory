```python
import os
from transformers import pipeline
import torch
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores.utils import DistanceStrategy
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from langchain.chains import RetrievalQA
from langchain_community.llms import HuggingFaceEndpoint
# from ragatouille import RAGPretrainedModel
from typing import List, Tuple

class Document:
    def __init__(self, page_content: str, metadata: dict):
        self.page_content = page_content
        self.metadata = metadata

class DocumentProcessor:
    def __init__(self):
        self.embedding_model_name = "thenlper/gte-small"
        self.embedding_model = HuggingFaceEmbeddings(
            model_name=self.embedding_model_name,
            multi_process=True,
            model_kwargs={"device": device},
            encode_kwargs={"normalize_embeddings": True}
        )
        self.reader_model_name = "HuggingFaceH4/zephyr-7b-beta"
        self.tokenizer = AutoTokenizer.from_pretrained(self.reader_model_name)
        self.reader_model = self._load_reader_model()
        # self.reranker = RAGPretrainedModel.from_pretrained("colbert-ir/colbertv2.0")
        self.knowledge_vector_database = None  # Will be initialized later

    def _load_reader_model(self):
        llm = HuggingFaceEndpoint(
        repo_id=self.reader_model_name,
        # max_length=128,
        top_k=10,
        temperature=0.5,
        # repetition_penalty=1.03,
        # add_to_git_credential=True,
        # load_in_8bit=True,
        huggingfacehub_api_token="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"
             )
        return llm

    def initialize_vector_database(self, path: str):
        self.knowledge_vector_database = FAISS.load_local(path, self.embedding_model, allow_dangerous_deserialization=True)


    def answer_question(self, question: str) -> Tuple[str, List[Document]]:
          relevant_docs = self.knowledge_vector_database.similarity_search(query=question)
          relevant_docs = [Document(doc.page_content, doc.metadata) for doc in relevant_docs]

          # Extract context from retrieved documents
          context = "\n".join([doc.page_content for doc in relevant_docs])

          # Build prompt with alternating roles

          final_prompt = self.tokenizer.apply_chat_template(
            [
                {"role": "system", "content": """Using the information contained in the context,
                                                give a comprehensive answer to the question.
                                          Respond in the langauage the question was provided.
                                          Respond only to the question asked, response should be concise and relevant to the question.
                                        If the answer cannot be deduced from the context,  ask for externals sources."""},  # You can fill this with appropriate system messages
                {"role": "user", "content": f"Context:\n{context}\n---\nNow here is the question you need to answer.\n\nQuestion: {question}"}
            ],
            tokenize=False,
            add_generation_prompt=True
        )

          print("=> Generating answer...")
          answer = self.reader_model(final_prompt)
          return {"answer": answer, "relevant_docs": context}
```

#QA avec Mistral

```python
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

class QA:
    def __init__(self):
        self.embedding_model_name = "thenlper/gte-small"
        self.embedding_model = HuggingFaceEmbeddings(
            model_name=self.embedding_model_name,
            multi_process=True,
            model_kwargs={"device": device},
            encode_kwargs={"normalize_embeddings": True}
        )
        self.knowledge_vector_database = None  # Will be initialized later
        self.reader_model = self._load_reader_model()
    def _load_reader_model(self, llm_model = "mistralai/Mixtral-8x7B-Instruct-v0.1", top_k = 10, temperature = 0.2):
         llm = HuggingFaceEndpoint(
              repo_id=llm_model,
              # max_length=128,
              top_k=top_k,
              temperature=temperature,
              # repetition_penalty=1.03,
              # add_to_git_credential=True,
              # load_in_8bit=True,
              huggingfacehub_api_token="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"
                  )
         return llm


    def initialize_vector_database(self, path: str):
        self.knowledge_vector_database = FAISS.load_local(path, self.embedding_model, allow_dangerous_deserialization=True)
    def answer(self, question: str):

      retriever = self.knowledge_vector_database.as_retriever(search_kwargs={"k": 10}) # Your retrieve

      system_prompt = (
          """You are a chatbot having a conversation with a human.

                 Given the following extracted parts of a long document and a question, create a final answer."""
          "Context: {context}"
      )
      prompt = ChatPromptTemplate.from_messages(
          [
              ("system", system_prompt),
              ("human", "{input}"),
          ]
      )
      question_answer_chain = create_stuff_documents_chain(self.reader_model, prompt)
      chain = create_retrieval_chain(retriever, question_answer_chain)

      return {"answer" : chain.invoke({"input": question })["answer"]}
```

#Mistral with memory

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.llms import HuggingFaceEndpoint
from langchain.memory import ConversationBufferMemory
# from langchain.chains import load_qa_chain
# from langchain.document_loaders import Document

class QA:
    def __init__(self):
        self.embedding_model_name = "thenlper/gte-small"
        self.embedding_model = HuggingFaceEmbeddings(
            model_name=self.embedding_model_name,
            multi_process=True,
            model_kwargs={"device": device},
            encode_kwargs={"normalize_embeddings": True}
        )
        self.knowledge_vector_database = None
        self.reader_model = None

    def initialise_qa_chain(self, temperature=0.2, top_k=10, llm_model="mistralai/Mixtral-8x7B-Instruct-v0.1",
                            template="""You are a chatbot having a conversation with a human.

        Given the following extracted parts of a long document and a question, create a final answer.

        {context}

        {chat_history}
        query: {human_input}
        """):

        llm = HuggingFaceEndpoint(
            repo_id=llm_model,
            top_k=top_k,
            temperature=temperature,
            # load_in_8bit=True,
            huggingfacehub_api_token="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"
        )

        prompt = PromptTemplate(
            input_variables=["chat_history", "human_input", "context"], template=template
        )
        memory = ConversationBufferMemory(memory_key="chat_history", input_key="human_input")
        chain = load_qa_chain(
            llm, chain_type="stuff", memory=memory, prompt=prompt
        )

        self.reader_model = chain

    def initialize_vector_database(self, path: str):
        self.knowledge_vector_database = FAISS.load_local(path, self.embedding_model, allow_dangerous_deserialization=True)

    def answer(self, question: str):
        docs = self.knowledge_vector_database.similarity_search(query=question, k=10)
        answer = self.reader_model({"input_documents": docs, "human_input": question}, return_only_outputs=True)
        return {"answer" : answer["output_text"]}

# Example usage
QA_instance = QA()
path = "/content/drive/MyDrive/kertys_db"
QA_instance.initialize_vector_database(path)
QA_instance.initialise_qa_chain()
response = QA_instance.answer("What did the president say about Justice Breyer")
print(response)
```

```python
from langchain_community.document_loaders import TextLoader
# from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores import FAISS
embedding_model_name = "thenlper/gte-small"
embedding_model = HuggingFaceEmbeddings(
            model_name=embedding_model_name,
            multi_process=True,
            model_kwargs={"device": device},
            encode_kwargs={"normalize_embeddings": True})
# Load the document, split it into chunks, embed each chunk and load it into the vector store.
# raw_documents = TextLoader('../../../state_of_the_union.txt').load()
text_splitter = CharacterTextSplitter(chunk_size=512, chunk_overlap=20)
documents = text_splitter.split_documents(data)
db = FAISS.from_documents(documents, embedding_model)
```
